## Python에서 HTTP로 파일 다운로드 하는 예제 프로그램

# 1.먼저 HTTP Server 실행 하자

```python
import socket

target_host = "192.168.0.62" 
target_URI = "/audio2.wav" 
#target_URI = "/sys_nv.dat" 
target_port = 8080  # create a socket object 

client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)  
 
# connect the client 
client.connect((target_host,target_port))  
 
# send some data 
#request = "GET %s HTTP/1.1\r\nHost:%s\r\n\r\n" %( target_URI,target_host)
#request = "GET %s HTTP/1.1\r\n\r\n" %target_URI
request = "GET %s HTTP/1.1\r\nUser-Agent: snk\r\nAccept: */*\r\nConnection: Close\r\n\r\n" %target_URI
client.send(request.encode())  
 
# receive some data 
response = client.recv(4096)  
http_response = repr(response)
http_response_len = len(http_response)
print(http_response)
print('length is',http_response_len)

response = client.recv(4096)  
http_response = repr(response)
http_response_len = len(http_response)
print(http_response)
print('length is',http_response_len)
```
